```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time


# Функция для парсинга сайта
def parse_website(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    # Пример извлечения данных (зависит от структуры сайта)
    titles = soup.find_all("h2")  # Предположим, что заголовки находятся в тегах <h2>
    for title in titles:
        print(title.text)


# Функция для автоматизации нажатия кнопок
def automate_button_click(url, button_text):
    # Настройка драйвера (например, для Chrome)
    driver = webdriver.Chrome()

    try:
        driver.get(url)
        time.sleep(2)  # Ожидание загрузки страницы

        # Поиск кнопки по тексту
        button = driver.find_element(
            By.XPATH, f"//button[contains(text(), '{button_text}')]"
        )
        button.click()
        time.sleep(2)  # Ожидание после нажатия

    finally:
        driver.quit()


# Основная функция
if __name__ == "__main__":
    url = "https://example.com"  # Замените на URL вашего сайта
    button_text = "Click Me"  # Замените на текст кнопки, которую нужно нажать

    # Парсинг сайта
    parse_website(url)

    # Автоматизация нажатия кнопки
    automate_button_click(url, button_text)
```
